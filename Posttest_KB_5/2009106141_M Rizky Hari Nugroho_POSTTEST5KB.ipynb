# Import Library

```python
import pandas as pd

# Data Preprocessing
from sklearn.model_selection import train_test_split # Data Split
from sklearn.preprocessing import StandardScaler # Standardisasi
from sklearn.preprocessing import MinMaxScaler # Normalisasi

# Algoritma
from sklearn.neighbors import KNeighborsClassifier # KNN
from sklearn.naive_bayes import GaussianNB # Naive Bayes
from sklearn.tree import DecisionTreeClassifier # Decision Tree
```

# Data Split

```python
data = df[['store_sales(in millions)','store_cost(in millions)','unit_sales(in millions)','cost']]
label = df['food_category']

data_train, data_test, label_train, label_test = train_test_split(data, label, test_size=.3)
```

```python
print(f'panjang data training {len(data_train)}',
     f'panjang label training {len(label_train)}',
     f'panjang data testing {len(data_test)}',
     f'panjang label testing {len(label_test)}',
     sep='\n')
```

# Normalisasi

```python
#Normalisasi
df_copy = df.copy()

scaler = MinMaxScaler()

normalize_dataset = scaler.fit_transform(df_copy[['store_sales(in millions)',
                                                  'store_cost(in millions)',
                                                  'unit_sales(in millions)',
                                                  'cost']])

normalize_dataset = pd.DataFrame(normalize_dataset)
normalize_dataset.rename(columns={ 0 :'store_sales(in millions)', 
                                  1 : 'store_cost(in millions)', 
                                  2 : 'unit_sales(in millions)', 
                                  3 : 'cost'}, inplace=True)

normalize_dataset.head(5)
```

# Standarisasi

```python
#Standarisasi
df_copy.drop('food_category', axis=1, inplace=True)

standart_std = StandardScaler()

standarisasi_data = standart_std.fit_transform(df_copy[['store_sales(in millions)',
                                                  'store_cost(in millions)',
                                                  'unit_sales(in millions)', 
                                                  'cost']])
standarisasi_data = pd.DataFrame(standarisasi_data)
standarisasi_data.rename(columns={  0 :'store_sales(in millions)', 
                            1 : 'store_cost(in millions)', 
                            2 : 'unit_sales(in millions)', 
                            3 : 'cost'}, inplace=True)
standarisasi_data
```

# Klasifikasi K-Nearest Neighbor

```python
knn_classifier = KNeighborsClassifier(n_neighbors = 5)
knn_classifier.fit(data_train, label_train)
```

# Klasifikasi Gaussian Naive Bayes

```python
gaussianNB_classifier = GaussianNB()
gaussianNB_classifier.fit(data_train, label_train)
```

# Klasifikasi Decision Tree

```python
decision_tree_classifier = DecisionTreeClassifier()
decision_tree_classifier.fit(data_train, label_train)
```

# SKOR

```python
print(f'Skor Model K-Nearest Neighbor dengan k=5\t: {knn_classifier.score(data_test, label_test)}',
      f'Skor Model Gaussian Naive Bayes\t\t\t: {gaussianNB_classifier.score(data_test, label_test)}',     
      f'Skor Model Decision Tree\t\t\t: {decision_tree_classifier.score(data_test, label_test)}',
      sep = '\n')
```
