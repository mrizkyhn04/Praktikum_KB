# Import Library

```python
import pandas as pd
import numpy as np

# DATA SPLIT
# Testing dan Training data
from sklearn.model_selection import train_test_split 

# DATA TRANSFORMING
# Standarisasi
from sklearn.preprocessing import StandardScaler 

# Normalisasi
from sklearn.preprocessing import MinMaxScaler

# DATA CLEANING
# Subtitusi nilai
from sklearn.impute import SimpleImputer 

# ENCODING
# Onehot
from sklearn.preprocessing import OneHotEncoder
```

# Membaca File CSV Disimpan di Dalam Variabel

```python
df = pd.read_csv('D:\Kuliah\Semester 5\Prak KB\cost_prediction.csv')
df.head(5)
```

# DATA SPLIT
Bagi dataset menjadi training set dan testing set dengan proporsi 70:30

```python
data = df[['store_sales(in millions)','store_cost(in millions)','unit_sales(in millions)','cost']]
label = df['food_category']

data_train, data_test, label_train, label_test = train_test_split(data, label, test_size=.3)
```

```python
print(f'panjang data training {len(data_train)}',
     f'panjang label training {len(label_train)}',
     f'panjang data testing {len(data_test)}',
     f'panjang label testing {len(label_test)}',
     sep='\n')
```

# DATA TRANSFORMASI
Lakukan normalisasi data pada salah satu attribute menggunakan Min Max scaler (buatlah copy dataset terlebih dahulu)

```python
#Normalisasi
df_copy = df.copy()

scaler = MinMaxScaler()

normalize_dataset = scaler.fit_transform(df_copy[['store_sales(in millions)',
                                                  'store_cost(in millions)',
                                                  'unit_sales(in millions)',
                                                  'cost']])
```

```python
normalize_dataset = pd.DataFrame(normalize_dataset)
normalize_dataset.rename(columns={ 0 :'store_sales(in millions)', 
                                  1 : 'store_cost(in millions)', 
                                  2 : 'unit_sales(in millions)', 
                                  3 : 'cost'}, inplace=True)

normalize_dataset.head(5)
```

# Lakukan standarisasi pada dataset 
(buatlah copy dataset terlebih dahulu)

```python
#Standarisasi
df_copy.drop('food_category', axis=1, inplace=True)

standart_std = StandardScaler()

#SEBELUM
scaled_data_cost = standart_std.fit_transform(df[['store_sales(in millions)',
                                                  'store_cost(in millions)',
                                                  'unit_sales(in millions)',
                                                  'cost']])
scaled_data_cost = pd.DataFrame(scaled_data_cost)
scaled_data_cost.rename(columns={ 0 :'store_sales(in millions)', 
                                1 : 'store_cost(in millions)', 
                                2 : 'unit_sales(in millions)', 
                                3 : 'cost'}, inplace=True)

#SESUDAH
scaled_data = standart_std.fit_transform(df_copy[['store_sales(in millions)',
                                                  'store_cost(in millions)',
                                                  'unit_sales(in millions)', 
                                                  'cost']])
scaled_data = pd.DataFrame(scaled_data)
scaled_data.rename(columns={  0 :'store_sales(in millions)', 
                            1 : 'store_cost(in millions)', 
                            2 : 'unit_sales(in millions)', 
                            3 : 'cost'}, inplace=True)
```

```python
print('Dataset sebelum scalling:',  # SEBELUM
      df.drop('food_category', axis=1).head(6),'\n'
      f'Nilai standar deviasi:\n{np.std(scaled_data_cost)}',
      sep='\n', end='\n\n')

print('Dataset setelah scalling:',  # SESUDAH
      scaled_data[:6],
      f'Nilai standar deviasi:\n{np.std(scaled_data)}',
      sep='\n')
```
