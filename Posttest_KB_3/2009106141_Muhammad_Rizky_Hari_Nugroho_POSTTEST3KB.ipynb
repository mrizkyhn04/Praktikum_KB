# Membuat Diagram Batang

```python
plt.figure(figsize = (9,4))
df['media_type'].value_counts().plot.bar()
plt.title('Tipe Media Promosi',fontsize = 20 , fontweight = 'bold')
plt.xlabel('Jenis Media',fontsize = 15)
plt.ylabel('Jumlah Promosi',fontsize = 15)
plt.show()
```

Dari Grafik Batang di atas diperoleh kesimpulan untuk penggunaan jenis media promosi terbanyak adalah promosi dengan media Koran Harian (Daily Paper) dan Radio dengan jumlah lebih dari 6000 pengguna. Dan untuk  penggunaan jenis media promosi paling sedikit adalah promosi dengan media Koran tiap Minggu (Sunday Paper), Radio dan TV dengan jumlah hanya lebih dari 3000 pengguna.

# Membuat Diagram Pie

```python
plt.figure(figsize = (4,3))
plt.pie(df['sales_country'].value_counts(), 
        colors = ['blue','green','orange'], 
        labels = ['USA','Mexico','Canada'],
        explode = (0,0,0.1),
        startangle = 45, 
        autopct = '%.1f%%')
plt.title('Penjualan Di Negara', fontweight = 'bold')
plt.legend(loc='lower left')
plt.show()
```

Dari Grafik Pie di atas diperoleh kesimpulan untuk penjualan terbanyak terletak di Negara USA dengan besar presentase yaitu 64,4% dari total keseluruhan penjualan. Dan untuk Mexico presentase penjualan sebesar 29,1% serta Canada merupakan negara dengan presentase penjualan terkecil yaitu hanya sebesar 6,6% dari jumlah keseluruhan penjualan.

# Membuat Grafik Heatmap

```python
sns.heatmap(data = df[['store_sales(in millions)', 'store_cost(in millions)', 'unit_sales(in millions)']].corr(), cmap='Reds')
plt.title('Korelasi antar atribut Penjualan Toko, Pengeluaran Toko, \n dan Unit Terjual\n', fontweight = 'bold')
plt.show()
```

Dari Grafik Heatmap di atas dapat diambil kesimpulan korelasi terkuat yaitu berada pada korelasi antar atribut itu sendiri seperti korelasi Penjualan Toko (store_sales) terhadap Penjualan Toko (store_sales) ditandai dengan warna yang paling gelap dan untuk korelasi terlebmah terdapat pada korelasi Pengeluaran Toko (store_cost) terhadap Unit Penjualan (unit_sales) ditandai dengan warna yang paling pudar.
