# Dictionary

```python
dictionary = {
    "Jumlah Record" : 0,
    "Nama Attribute" : [],
    "Jumlah Attribute Tipe Objek" : 0,
    "Jumlah Attribute Tipe Angka" : 0,
    "Jumlah Attribute Dengan Nilai Null" : 0
}
```

# Jumlah Record

```python
dictionary["Jumlah Record"] = dataframe.shape[0]
```

# Nama-nama Attribute

```python
dictionary["Nama Attribute"] = list(dataframe.columns)
```

# Jumlah Attribute Bertipe Objek 

```python
dictionary["Jumlah Attribute Tipe Objek"] = len(dataframe.select_dtypes(include = ["object"]).columns)
```

# Jumlah Attribute Bertipe Angka

```python
dictionary["Jumlah Attribute Tipe Angka"] = len(dataframe.select_dtypes(exclude = ["object"]).columns)
```

# Jumlah Attribute dengan Nilai Null

```python
null_value = 0
for value in dict(dataframe.isnull().sum()).values():
  if (value > 0):
    null_value += value
dictionary["Jumlah Attribute Dengan Nilai Null"] = null_value
```

# Menampilkan Isi Dictionary

```python
for key, value in dictionary.items():
  if (len(key) == 1):
    print(f"{key} : {value}")
  elif (len(key) == 14):
    print(f"{key}  :")
    for index, i in enumerate(value):
      print(f"{index + 1}. {i}")
    print(f"\n")
  elif (len(key) == 27):
    print(f"{key}\t\t: {value}")
  else:
    print(f"{key}\t: {value}")
```
